```python
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
```

```python
data = pd.read_excel(r"C:\Users\Xi-Jie Hou\Desktop\load2.xlsx")
```

```python
d1=[(ix, k, v) for ix, row in data.iterrows() for k, v in row.items()]
```

```python
d2=[]
for i in range(len(d1)):
    if not np.isnan(d1[i][2]):
        d2.append(d1[i])
```

```python
print(d2)
```

```python
X,Y,Z = zip(*d2)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
surf = ax.plot_trisurf(X, Y, Z, cmap=cm.jet, linewidth=0.1)
fig.set_size_inches(8,6)
plt.xlabel('Label Privacy Budget   $log_{10}(\epsilon_c)$', fontsize=15)
plt.ylabel('Image Privacy Budget   $log_{10}(\epsilon_x)$', fontsize=15)
ax.set_zlabel('Test Accuracy', fontsize=15)
ax.view_init(azim=345)
fig.savefig(r"C:\Users\Xi-Jie Hou\Desktop\345.png", dpi=300)
plt.show()
```
